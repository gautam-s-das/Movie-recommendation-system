```python
import pandas as pd
import numpy as np
#import warnings 
#warnings.filterwarnings('ignore')
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies=movies.merge(credits,on='title')
```

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew
movies=movies[['movie_id','title','overview','keywords','genres','cast','crew']]
```

```python
#droping null values
movies.dropna(inplace=True)
```

```python
#checking duplicate values
movies.duplicated().sum()
```

```python
#taking the first row of genres field 
movies.iloc[0].genres
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#['action','adventure','fantacy','scifi']
#converting the string into list
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
#apply the convert fun to genres
movies['genres']=movies['genres'].apply(convert)
```

```python
#genres fiels has changed to ['action','adventure','fantacy','scifi']
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#taking only first 3 cast members from cast team
import ast
def convert3(obj):
    counter=0
    L=[]
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#fetching only director from crew column where job==director
import ast
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
#converting overview string into list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#creating a new tag column and adding overview,genres,keywords,cast,crew
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
#creating a new dataframe
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'].apply(lambda x:x.lower())
```

```python
#applying natural language tool kit
import nltk
```

```python
#Stemmers remove morphological affixes from words, leaving only the word stem.
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return" ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
# convert text into vectors
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
#It returns the list of feature names (i.e., vocabulary terms) that the CountVectorizer has learned from the text data.
#In simpler terms, it tells you what words (tokens) it has extracted and used as features when converting text into numbers.
cv.get_feature_names_out()
```

```python
#measure similarity between two vectors.
# 1 if they point in the same direction (very similar),
# 0 if they are orthogonal (completely different),
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# similarity[i][j] = similarity between document i and document j
similarity=cosine_similarity(vectors)
```

```python
# new_df: DataFrame containing movie data

# similarity: cosine similarity matrix (usually shape: n_movies × n_movies)

# Given a movie title:

# Finds its index

# Fetches cosine similarity scores for that movie

# Sorts them to find the top 5 most similar movies (excluding itself)

# Prints their indices
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    print(f"\nTop 5 movies similar to '{movie}':")
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
